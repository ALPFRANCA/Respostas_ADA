1) Utilizando o DataFrame importado anteriormente (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/alunos3.csv', sep=';', decimal=',')
df = df.astype({'Prova_1': float, 'Prova_2': float, 'Prova_3': float, 'Prova_4': float})
df['Media_provas'] = df[['Prova_1','Prova_2','Prova_3','Prova_4']].mean(axis=1)
df
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
df.loc[df['Media_provas'].idxmax()]
```

```python
df.loc[df['Media_provas'].idxmin()]
```

3) Agora una este dataframe com o cadastro_alunos.xlsx

```python
ca = pd.read_excel('/content/cadastro_alunos.xlsx')
ca
```

```python
df_ca = pd.merge(df, ca, on='RA')
df_ca
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df_ca.groupby('Sexo')['Media_provas'].describe()
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
df_ca.loc[df_ca['Media_provas'] >= 7]['Idade'].mean()
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
df_ca.groupby('Cidade')['Media_provas'].mean()
```
