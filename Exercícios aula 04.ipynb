```python
import numpy as np
import pandas as pd
```

```python
#1 - Crie uma Serie com elementos A, B, C e com os indexes primeiro, segundo, terceiro

serie = pd.Series(['A', 'B', 'C'], index=['primeiro', 'segundo', 'terceiro'])
serie
```

```python
#2 - Agora mostre o primeiro elemento da Serie anterior

serie.iloc[0]
```

```python
#3 - Baseado na Serie X = pd.Series([1,2,3,4,5], index=['first','second','third','forth','fifth']) - Transforme em float

X = pd.Series([1,2,3,4,5], index=['first','second','third','forth','fifth'])
X = X.astype(float)
X
```

```python
#4 - Mude a ordem da mesma Serie utilizada no Exercicio 3

X = X.reindex(index=['second','first','third','fifth','forth'])
X
```

```python
#5 - Organize os valores da Serie X = pd.Series([4,2,5,1,3], index=['forth','second','fifth','first','third'])

Y = pd.Series([4,2,5,1,3], index=['forth','second','fifth','first','third'])
Y = Y.reindex(index=['first','second','third','forth','fifth'])
Y
```

```python
#6 - Add 5 para todos os elementos da Serie X = pd.Series([1,2,3,4,5])

Z = pd.Series([1,2,3,4,5])
Z = Z + 5
Z
```

```python
#7 - Crie uma m√°scara para mostrar somente elementos negativos K = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

K = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])
mask = K < 0
K[mask]
```

```python
#8 - Colete todos os elementos negativos K = pd.Series([-1,2,0,-4,5,6,0,0,-9,10])

W = pd.Series(K[mask])
W
```
